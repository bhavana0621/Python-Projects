```python
class Bus:
    def __init__(self, bus_no, total_seats):
        self.bus_no = bus_no
        self.total_seats = total_seats
        self.seats = ["Empty"] * total_seats

    def display_info(self):
        print(f"\nBus No: {self.bus_no}")
        print(f"Total Seats: {self.total_seats}")

    def show_seats(self):
        print("\nSeat Status:")
        for i in range(len(self.seats)):
            print(f"Seat {i+1}: {self.seats[i]}")

    def reserve_seat(self, seat_no, name):
        if self.seats[seat_no - 1] == "Empty":
            self.seats[seat_no - 1] = name
            print("✅ Seat booked successfully!")
        else:
            print("❌ Seat already booked!")


class ReservationSystem:
    def __init__(self):
        self.buses = []

    def add_bus(self):
        bus_no = input("Enter Bus Number: ")
        seats = int(input("Enter total seats: "))
        bus = Bus(bus_no, seats)
        self.buses.append(bus)
        print("✅ Bus added successfully!")

    def view_buses(self):
        if not self.buses:
            print("No buses available!")
            return
        for bus in self.buses:
            bus.display_info()

    def select_bus(self):
        bus_no = input("Enter Bus Number: ")
        for bus in self.buses:
            if bus.bus_no == bus_no:
                return bus
        print("❌ Bus not found!")
        return None

    def reserve(self):
        bus = self.select_bus()
        if bus:
            bus.show_seats()
            seat_no = int(input("Enter seat number: "))
            name = input("Enter passenger name: ")
            bus.reserve_seat(seat_no, name)

    def show_reservations(self):
        bus = self.select_bus()
        if bus:
            bus.show_seats()


def main():
    system = ReservationSystem()

    while True:
        print("\n--- Bus Reservation System ---")
        print("1. Add Bus")
        print("2. View Buses")
        print("3. Reserve Seat")
        print("4. Show Reservations")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            system.add_bus()
        elif choice == "2":
            system.view_buses()
        elif choice == "3":
            system.reserve()
        elif choice == "4":
            system.show_reservations()
        elif choice == "5":
            print("Thank you!")
            break
        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
