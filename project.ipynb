```python
# Grocery Billing System

items = ["Rice", "Wheat", "Sugar", "Oil", "Milk"]
prices = [50, 40, 45, 120, 30]

cart_items = []
cart_prices = []

def show_items():
    print("\nAvailable Items:")
    for i in range(len(items)):
        print(i+1, items[i], "- Rs.", prices[i])

def add_item():
    choice = int(input("Enter item number to add: "))
    quantity = int(input("Enter quantity: "))

    if choice >= 1 and choice <= len(items):
        cart_items.append(items[choice-1])
        cart_prices.append(prices[choice-1] * quantity)
        print("Item added to cart")
    else:
        print("Invalid choice")

def generate_bill():
    total = sum(cart_prices)

    print("\n----- BILL -----")
    for i in range(len(cart_items)):
        print(cart_items[i], "- Rs.", cart_prices[i])

    if total >= 500:
        discount = total * 0.10
        print("Discount: Rs.", discount)
        total = total - discount
    else:
        print("No Discount")

    print("Total Amount to Pay: Rs.", total)
    print("----------------")

def main():
    while True:
        print("\n----GROCERY BILLING SYSTEM -----")
        print("1. Show Items")
        print("2. Add Item")
        print("3. Generate Bill")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            show_items()
        elif choice == "2":
            add_item()
        elif choice == "3":
            generate_bill()
        elif choice == "4":
            print("Thank you for shopping!")
            break
        else:
            print("Invalid option")

main()
```
